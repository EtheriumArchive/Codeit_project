```python
import pandas as pd
import numpy as np
import koreanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

### 신고가 몰린 날짜 추이

```python
timelinereport_created_at = pd.read_csv('./timelinereport_created_at.csv')

timelinereport_created_at['created_at'] = pd.to_datetime(timelinereport_created_at['created_at'])

ax = sns.lineplot(data=timelinereport_created_at, x='created_at', y='counts')
ax.tick_params(axis='x', labelrotation=45)
ax.set_title('신고가 몰린 날짜')
plt.show()
```

### 신고 많이 하는 시간대

```python
report_times = pd.read_csv('./report_times.csv')
# vote_times['answer_updated_at'] = pd.to_datetime(vote_times['answer_updated_at'])

ax = sns.lineplot(data=report_times, x='created_at', y='counts')
ax.set_xticks(range(0,24))
ax.set_title('신고가 몰린 시간대')
plt.show()
```

### 투표 많이 하는 시간대

```python
vote_times = pd.read_csv('./vote_times.csv')
# vote_times['answer_updated_at'] = pd.to_datetime(vote_times['answer_updated_at'])
vote_times['answer_updated_at'] = vote_times['answer_updated_at'] + 9
ax = sns.lineplot(data=vote_times, x='answer_updated_at', y='counts')
ax.set_xticks(range(0,24))
ax.set_title('투표 답변 완료한 시간')
plt.show()
```

### UTC 시간대 변경 (KST)

```python
vote_times['answer_updated_at'] = vote_times['answer_updated_at'] + 9
```

```python
vote_times['answer_updated_at'] = vote_times['answer_updated_at'].apply(lambda x: x if x < 24 else x-24)
```

```python
ax = sns.lineplot(data=vote_times, x='answer_updated_at', y='counts')
ax.set_xticks(range(0,24))
ax.set_title('투표 답변 완료한 시간')
plt.show()
```
